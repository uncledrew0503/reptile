```python
import requests
import re
import time
from bs4 import BeautifulSoup
def getHTMLText(url):
    try:
        hd={'user-agent':'Chrome/10'}
        r=requests.request('POST',url,headers=hd)
        r.raise_for_status()
        r.encoding=r.apparent_encoding
        return r.text
    except:
        return ''
def parsePage(list1,html):#获取商品的名称和价格
    try:
        plt1=re.findall('<a title="[u4e00-\u9fa5\w\(\)\s]+网签',html)#小区名称
        plt2=re.findall('istrict" title="[u4e00-\u9fa5]+小区',html)#区
        plt3=re.findall('class="bizcircle" title="[u4e00-\u9fa5]+小区',html)#县
        plt4=re.findall('<span>[0-9]+</span>元/m',html)#均价
        plt5=re.findall('90天成交[0-9]+套',html)#三个月成交数量
        plt6=re.findall('<span>[0-9]+</span>套',html)#在售数量
        for i in range(len(plt1)):
            ilt=[]
            ilt.append(plt1[i][10:-2])
            ilt.append(plt2[i][16:-2])
            ilt.append(plt3[i][25:-2])
            ilt.append(plt4[i][6:-10])
            ilt.append(plt5[i][5:-1])
            ilt.append(plt6[i][6:-8])
            list1.append(ilt)
    except: #防止遇到异常情况退出
        print("")
# def printGoodList(ilt):
#     tplt='{:^25}\t{:^10}\t{:^10}\t{:^10}\t{:^10}\t{:^10}'
#     print(tplt.format('位置','成交价格（万）','挂牌价格（万）','户型','面积（平米）','成交周期',chr(12288)))
#     count=0
#     for g in ilt:
#         count=count+1
#         print(tplt.format(count,g[0],g[1]))

def main(cond_num):
    depth=100
    start_url='https://sh.lianjia.com/xiaoqu/'
    infolist=[]
    for i in range(1,depth+1):
        #time.sleep(1)
        try:
            info=[]
            url=start_url+'pg'+str(i)+'p'+str(cond_num)
            html=getHTMLText(url)
            parsePage(info,html)
            infolist=infolist+info
        except:
            continue
    return infolist
```
