```python
import requests
import re
import time
def getHTMLText(url):
    try:
        hd={'user-agent':'Chrome/10'}
        r=requests.request('POST',url,headers=hd)
        r.raise_for_status()
        r.encoding=r.apparent_encoding
        return r.text
    except:
        return ''
def parsePage(ilt,html):#获取商品的名称和价格
    try:
        plt1=re.findall('target="_blank">[u4e00-\u9fa5\s\(\)\.]*平米</a></div>',html)#获取位置、户型、面积的数据
        plt2=re.findall('成交周期[0-9]*天',html)#获取成交周期的数据
        plt3=re.findall('[0-9.]+</span>万',html)#获取成交价格的数据
        plt4=re.findall('挂牌[0-9.]*万',html)#获取挂牌价格的数据
        for i in range(len(plt1)):
            ilt.append(plt1[i][16:-10].split())
            ilt[i].append(plt2[i][4:-1])
            ilt[i].insert(1,plt3[i][:-8])
            ilt[i].insert(2,plt4[i][2:-1])
    except: #防止遇到异常情况退出
        print("")
# def printGoodList(ilt):
#     tplt='{:^25}\t{:^10}\t{:^10}\t{:^10}\t{:^10}\t{:^10}'
#     print(tplt.format('位置','成交价格（万）','挂牌价格（万）','户型','面积（平米）','成交周期',chr(12288)))
#     count=0
#     for g in ilt:
#         count=count+1
#         print(tplt.format(count,g[0],g[1]))

def main(begin,end):
    depth=100
    start_url='https://sh.lianjia.com/chengjiao'
    infolist=[]
    for i in range(1,depth+1):
        #time.sleep(1)
        try:
            info=[]
            url=start_url+'/pg'+str(i)+'bp'+str(begin)+'ep'+str(end)
            html=getHTMLText(url)
            parsePage(info,html)
            infolist=infolist+info
        except:
            continue
    return infolist
```

```python
import requests
from bs4 import BeautifulSoup
import bs4

def getHTMLText(url):
    try:
        r=requests.get(url,timeout=20)
        r.raise_for_status()
        r.encoding=r.apparent_encoding
        return r.text
    except:
        return ''

def fillUnivList(ulist,html):
    soup=BeautifulSoup(html, 'html.parser')
    for tr in soup.find('tbody').children:
        if isinstance(tr, bs4.element.Tag):
            tds=tr('td')
            ulist.append([tds[0].string,tds[1].string,tds[3].string])

def printUnivList(ulist,num):
    tplt='{0:^10}\t{1:{3}^10}\t{2:^10}'
    print(tplt.format('排名','学校名称','总分',chr(12288)))
    for i in range(num):
        u=ulist[i]
        print(tplt.format(u[0],u[1],u[2],chr(12288)))
    print('suc'+str(num))

def main():
    uinfo=[]
    url='http://www.zuihaodaxue.com/zuihaodaxuepaiming2019.html'
    html=getHTMLText(url)
    fillUnivList(uinfo,html)
    printUnivList(uinfo,20)
main()
```

```python
def getHTMLText(url):
    try:
        r=requests.get(url,timeout=20)
        r.raise_for_status()
        r.encoding=r.apparent_encoding
        return r.text
    except:
        return ''

def parsePage(ilt,html):#获取商品的名称和价格
    try:
        plt=re.findall(r'\"view_price\"\:\"[\d\.]*\"',html)
        tlt=re.findall(r'\"raw_title\"\:\".*?\"',html)
        for i in range(len(plt)):
            price=eval(plt[i].split(':')[1])
            title=eval(tlt[i].split(':')[1])
            ilt.append([price,title])
    except: #防止遇到异常情况退出
        print("")
def printGoodList(ilt):
    tplt="{:4}\t{:8}\t{:16}"
    print(tplt.format('序号','价格','商品名称'))
    count=0
    for g in ilt:
        count=count+1
        print(tplt.format(count,g[0],g[1]))

def main():
    goods='书包'
    depth=2
    start_url='https://s.taobao.com/search?q='+goods
    infolist=[]
    for i in range(depth):
        try:
            url=start_url+'&s='+str(44*i)
            html=getHTMLText(url)
            parsePage(infolist,html)
        except:
            continue
    printGoodList(infolist)
    print(infolist)
main()
```

```python
from urllib import request
from lxml import etree
#构造函数，抓取第i页信息
def crow(i):
    #  构造第i页的网址
    url='https://movie.douban.com/top250?start='+str(25*i)
    #  发送请求，获得返回的html代码并保存在变量html中
    html=request.urlopen(url).read().decode('utf-8')
    #将返回的字符串格式的html代码转换成xpath能处理的对象
    html=etree.HTML(html)
    #先定位到li标签，datas是一个包含25个li标签的list，就是包含25部电影信息的list
    datas = html.xpath('//ol[@class="grid_view"]/li')
    a=0
    for data in datas:
        data_title=data.xpath('div/div[2]/div[@class="hd"]/a/span[1]/text()')
        data_info=data.xpath('div/div[2]/div[@class="bd"]/p[1]/text()')
        data_quote=data.xpath('div/div[2]/div[@class="bd"]/p[2]/span/text()')
        data_score=data.xpath('div/div[2]/div[@class="bd"]/div/span[@class="rating_num"]/text()')
        data_num=data.xpath('div/div[2]/div[@class="bd"]/div/span[4]/text()')
        data_picurl=data.xpath('div/div[1]/a/img/@src')
        print("No: "+str(i*25+a+1))
        print(data_title)
        #保存电影信息到txt文件，下载封面图片
        with open('data.txt','a',encoding='utf-8')as f:
            #封面图片保存路径和文件名
            picname='C:/Users/Administrator/Desktop/爬虫/pics/'+str(i*25+a+1)+'.jpg'
            f.write("No: "+str(i*25+a+1)+'\n')
            f.write(data_title[0]+'\n')
            f.write(str(data_info[0]).strip()+'\n')
            f.write(str(data_info[1]).strip()+'\n')
            #因为发现有几部电影没有quote，所以这里加个判断，以免报错
            if data_quote:
                f.write(data_quote[0]+'\n')
            f.write(data_score[0]+'\n')
            f.write(data_num[0]+'\n')
            f.write('\n'*3)
            #下载封面图片到本地，路径为picname
            request.urlretrieve(data_picurl[0],filename=picname)
        a+=1
for i in range(10):
    crow(i)
```
